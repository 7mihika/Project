```python
# Missing Data
```

```python
messy = pd.DataFrame({
    "produc":["pen","Book","Bag","Lamp","Mug"],
    "price":[20,np.nan,850,1200,np.nan],
    "rating":[4.1,4,np.nan,4.5,3.9]
})
```

```python
messy
```

```python
messy.isna().sum()
```

```python
filled = messy.copy()
```

```python
filled["price"] = filled["price"].fillna(filled["price"].mean())
```

```python
filled["rating"] = filled["rating"].fillna(filled["rating"].median())
```

```python
print(filled)
```
